## Validate Pytorch VS Triton Kernel at the model level

A much more aggressive, input / output delta comparision for the model input / output between the various time mix kernel implementation
This test the model logits itself, for blank inputs

```python
# Configure the parent path to be the proj folder
import sys, os, torch, time
sys.path.append('../../')

# Import the model classes
from rwkv_block.v7_goose.model.rwkv7_goose_model import RWKV7GooseModel
from rwkv_block.v7_goose.model.rwkv7_goose_config_map import RWKV7GooseConfigMap

# File to load
MODEL_FILENAME="v7-1B4.pth"

# Run device, and run dtype to use
RUN_DEVICE="cpu"
RUN_DTYPE=torch.bfloat16

# Check for cuda device
if torch.cuda.is_available():
    RUN_DEVICE="cuda:0"

# Check if the reference weights exists
assert os.path.exists(f"./.model/{MODEL_FILENAME}"), "The reference weights does not exist. Please download it first (00-model-download.ipynb)"
abs_model_path = os.path.abspath(f"./.model/{MODEL_FILENAME}")

# Loads the model weights
model_weight = torch.load(f"./.model/{MODEL_FILENAME}", map_location='cpu', weights_only=True, mmap=True)

# Model filename
print(f"### Model filename: {MODEL_FILENAME}")

# Lets get the hidden_size, and setup the test module
hidden_size = model_weight['emb.weight'].shape[1]
print(f"### Model hidden_size: {hidden_size}")

# # List the model weights keys, and their shapes
# print(f"### model weights keys:")
# for key in model_weight:
#     print(f"{key}: {model_weight[key].shape} - {model_weight[key].dtype}")
```

```python
BATCH_SIZE=1
IN_TOKENS_LEN=16

# Get the pytorch model
py_model_config = RWKV7GooseConfigMap.from_model_state_dict(model_weight, device=RUN_DEVICE, dtype=RUN_DTYPE, tmix_backend="pytorch")
py_model = RWKV7GooseModel(py_model_config)
py_model.load_state_dict(model_weight)
print("### Pytorch Model Config:", py_model_config)

# Get the triton model
tr_model_config = RWKV7GooseConfigMap.from_model_state_dict(model_weight, device=RUN_DEVICE, dtype=RUN_DTYPE, tmix_backend="triton")
tr_model = RWKV7GooseModel(tr_model_config)
tr_model.load_state_dict(model_weight)
print("### Triton Model Config:", tr_model_config)

# Get the cuda model
cu_model_config = RWKV7GooseConfigMap.from_model_state_dict(model_weight, device=RUN_DEVICE, dtype=RUN_DTYPE, tmix_backend="cuda")
cu_model = RWKV7GooseModel(cu_model_config)
cu_model.load_state_dict(model_weight)
print("### CUDA Model Config:", cu_model_config)
```

```python
# Dummy run for cuda model
x_tokens = torch.zeros(BATCH_SIZE, IN_TOKENS_LEN, device=py_model.emb.weight.device, dtype=torch.int)

with torch.inference_mode():
    cu_out_logit, cu_out_state = cu_model.forward(x_tokens, cu_model.get_init_state(BATCH_SIZE))
```

```python
# Lets prepare the states accordingly
x_tokens = torch.zeros(BATCH_SIZE, IN_TOKENS_LEN, device=py_model.emb.weight.device, dtype=torch.int)

# Lets get the respective outputs
with torch.inference_mode():
    # Pytorch
    py_out_logit, py_out_state = py_model.forward(x_tokens, py_model.get_init_state(BATCH_SIZE))
    py_last_softmax_logit = torch.nn.functional.softmax(py_out_logit[-1][-1])
    print(f"### Pytorch Output x (Shape: {py_out_logit.shape}", "max:", py_out_logit.max().item(), "min:", py_out_logit.min().item(), ")")

    # ---
    print("---")

    # Triton
    tr_out_logit, tr_out_state = tr_model.forward(x_tokens, tr_model.get_init_state(BATCH_SIZE))
    tr_last_softmax_logit = torch.nn.functional.softmax(tr_out_logit[-1][-1])
    print(f"### Triton Output x (Shape: {py_out_logit.shape}", "max:", tr_out_logit.max().item(), "min:", tr_out_logit.min().item(), ")")

    # Compute the delta logit
    delta_tr_logit = (py_out_logit[-1][-1] - tr_out_logit[-1][-1]).view(128,-1).float().cpu().numpy()
    delta_tr_logit_abs_max = max(abs(delta_tr_logit.max()), abs(delta_tr_logit.min()))
    # Log the delta logit shape
    print(f"### Triton Delta Logit (Shape: {delta_tr_logit.shape}, max:", delta_tr_logit.max(), ", min:", delta_tr_logit.min(), ")")

    # Compute the delta softmax
    delta_last_softmax = (py_last_softmax_logit - tr_last_softmax_logit).view(128,-1).float().cpu().numpy()
    delta_last_softmax_abs_max = max(abs(delta_last_softmax.max()), abs(delta_last_softmax.min()))
    # Log the delta softmax shape
    print(f"### Triton Delta Softmax (Shape: {delta_last_softmax.shape}, max:", delta_last_softmax.max(), ", min:", delta_last_softmax.min(), ")")

    # ---
    print("---")

    # CUDA
    cu_out_logit, cu_out_state = cu_model.forward(x_tokens, cu_model.get_init_state(BATCH_SIZE))
    cu_last_softmax_logit = torch.nn.functional.softmax(cu_out_logit[-1][-1])
    print(f"### CUDA Output x (Shape: {cu_out_logit.shape}", "max:", cu_out_logit.max().item(), "min:", cu_out_logit.min().item(), ")")

    # Compute the delta logit
    delta_cu_logit = (py_out_logit[-1][-1] - cu_out_logit[-1][-1]).view(128,-1).float().cpu().numpy()
    delta_cu_logit_abs_max = max(abs(delta_cu_logit.max()), abs(delta_cu_logit.min()))
    # Log the delta logit shape
    print(f"### CUDA Delta Logit (Shape: {delta_cu_logit.shape}, max:", delta_cu_logit.max(), ", min:", delta_cu_logit.min(), ")")

    # Compute the delta softmax
    delta_last_softmax = (py_last_softmax_logit - cu_last_softmax_logit).view(128,-1).float().cpu().numpy()
    delta_last_softmax_abs_max = max(abs(delta_last_softmax.max()), abs(delta_last_softmax.min()))
    # Log the delta softmax shape
    print(f"### CUDA Delta Softmax (Shape: {delta_last_softmax.shape}, max:", delta_last_softmax.max(), ", min:", delta_last_softmax.min(), ")")
```

```python
# Plot the delta logit
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_tr_logit, cmap='seismic', aspect='auto')
im.set_clim(-delta_tr_logit_abs_max, delta_tr_logit_abs_max)
ax.set_title('Delta Logit (pytorch -> triton)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```

```python
# Plot the delta logit
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_last_softmax, cmap='seismic', aspect='auto')
im.set_clim(-delta_last_softmax_abs_max, delta_last_softmax_abs_max)
ax.set_title('Delta Softmax Logit (pytorch -> triton)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```

```python
os.environ["RWKV_V7_ON"] = '1' # ==> enable RWKV-7 mode
os.environ['RWKV_JIT_ON'] = '1' # '1' for better speed
os.environ["RWKV_CUDA_ON"] = '0' # '1' to compile CUDA kernel (10x faster), requires c++ compiler & cuda libraries

from rwkv.model import RWKV
from rwkv.utils import PIPELINE, PIPELINE_ARGS
import numpy

# Load the model
filename_without_ext = f"./.model/{MODEL_FILENAME}"[:-4]

# Get the reference output
ref_model = RWKV(model=filename_without_ext, strategy="cuda bf16")

# Refence token to use for forwarding
ref_tokens = [0] * IN_TOKENS_LEN

# Forward with IN_TOKENS_LEN tokens of zero
ref_out_logit, ref_out_state = ref_model.forward(ref_tokens, None, full_output=True)

# Get the delta pytorch / reference output
delta_ref_py_out_logit = (ref_out_logit[-1] - py_out_logit[-1][-1]).view(128,-1).float().cpu().numpy()
delta_ref_py_out_logit_abs_max = numpy.abs(delta_ref_py_out_logit).max()
# Log the delta shape and max value
print(f"### delta_ref_py_out_logit.shape: {delta_ref_py_out_logit.shape}, max: {delta_ref_py_out_logit.max()}, min: {delta_ref_py_out_logit.min()}")

# Get the delta triton / reference output
delta_ref_tr_out_logit = (ref_out_logit[-1] - tr_out_logit[-1][-1]).view(128,-1).float().cpu().numpy()
delta_ref_tr_out_logit_abs_max = numpy.abs(delta_ref_tr_out_logit).max()
# Log the delta shape and max value
print(f"### delta_ref_tr_out_logit.shape: {delta_ref_tr_out_logit.shape}, max: {delta_ref_tr_out_logit.max()}, min: {delta_ref_tr_out_logit.min()}")

# Build the delta softmax
ref_last_softmax_logit = torch.nn.functional.softmax(ref_out_logit[-1])

# Compute the delta softmax for pytorch
delta_ref_py_softmax = (ref_last_softmax_logit - py_last_softmax_logit).view(128,-1).float().cpu().numpy()
delta_ref_py_softmax_abs_max = numpy.abs(delta_ref_py_softmax).max()
# Log the softmax delta shape and max value
print(f"### delta_ref_py_softmax.shape: {delta_ref_py_softmax.shape}, max: {delta_ref_py_softmax.max()}, min: {delta_ref_py_softmax.min()}")

# Compute the delta softmax for triton
delta_ref_tr_softmax = (ref_last_softmax_logit - tr_last_softmax_logit).view(128,-1).float().cpu().numpy()
delta_ref_tr_softmax_abs_max = numpy.abs(delta_ref_tr_softmax).max()
# Log the softmax delta shape and max value
print(f"### delta_ref_tr_softmax.shape: {delta_ref_tr_softmax.shape}, max: {delta_ref_tr_softmax.max()}, min: {delta_ref_tr_softmax.min()}")

# For plotting
import matplotlib.pyplot as plt
```

```python
# Plot the delta logit
fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_ref_py_out_logit, cmap='seismic', aspect='auto')
im.set_clim(-delta_ref_py_out_logit_abs_max, delta_ref_py_out_logit_abs_max)
ax.set_title('Delta of Last Logit (reference -> pytorch)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```

```python
# Plot the delta logit
fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_ref_tr_out_logit, cmap='seismic', aspect='auto')
im.set_clim(-delta_ref_tr_out_logit_abs_max, delta_ref_tr_out_logit_abs_max)
plt.title('Delta of Last Logit (reference -> triton)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```

```python
# Plot the softmax delta
fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_ref_py_softmax, cmap='seismic', aspect='auto')
im.set_clim(-delta_ref_py_softmax_abs_max, delta_ref_py_softmax_abs_max)
ax.set_title('Delta of Last Softmax (reference - pytorch)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```

```python
# Plot the softmax delta
fig, ax = plt.subplots(figsize=(8, 6))  # Set a minimum figure size
im = ax.imshow(delta_ref_tr_softmax, cmap='seismic', aspect='auto')
im.set_clim(-delta_ref_tr_softmax_abs_max, delta_ref_tr_softmax_abs_max)
ax.set_title('Delta of Last Softmax (reference - triton)')
fig.colorbar(im, ax=ax)
fig.tight_layout()
plt.show()
```
